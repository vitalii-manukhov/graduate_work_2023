```python
import matplotlib.pyplot as plt

# Значения для первой выборки
fractions_1 = ['1', '1/2', '1/4', '1/8', '1/16', '1/32', '1/64']
values_1 = [0.89913, 0.836932, 0.813335, 0.823121, 0.774448, 0.717801, 0.681826]

# Значения для второй выборки
fractions_2 = ['1', '1/2', '1/4', '1/8', '1/16', '1/32', '1/64']
values_2 = [0.933337, 0.910273, 0.864923, 0.815335, 0.814242, 0.787981, 0.72637]

# Ширина столбцов
bar_width = 0.35

# Создание фигуры и осей
fig, ax = plt.subplots()

# Расположение столбцов для первой выборки
bar_positions_2 = np.arange(len(fractions_1))
ax.bar(bar_positions_2, values_2, width=bar_width, color='orange', label='Результат применения трансферного обучения')

# Расположение столбцов для второй выборки
bar_positions_1 = bar_positions_2 + bar_width
ax.bar(bar_positions_1, values_1, width=bar_width, color='blue', label='Без применения трансферного обучения')

# Настройка осей и заголовка
ax.set_xlabel('Количество данных')
ax.set_ylabel('AUC')
ax.set_title('Трансферное обучение')
ax.set_xticks(bar_positions_2 + bar_width / 2)
ax.set_xticklabels(fractions_1)

# Легенда
ax.legend(loc='upper left', bbox_to_anchor=(1, 1))

# Отзеркальвание графика
ax.invert_xaxis()

# Отображение графика
plt.show()
```
